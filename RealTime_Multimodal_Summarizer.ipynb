```python
import base64
import os
from langchain_core.messages import HumanMessage
```

```python
def encode_image(image_path):
    """Getting the base64 string"""
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
```

```python
fpath = '/content/extracted_data/figure-1-1.jpg'
encode_image(fpath)
```

```python
def image_summarize(img_base64, prompt):
    """Make image summary"""


    chat = model

    msg = chat.invoke(
        [
            HumanMessage(
                content=[
                    {"type": "text", "text": prompt},

                     {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{img_base64}"},
                    },
                ]
            )
        ]
    )
    return msg.content
```

```python
def generate_img_summaries(path):
    """
    Generate summaries and base64 encoded strings for images
    path: Path to list of .jpg files extracted by Unstructured
    """

    # Store base64 encoded images
    img_base64_list = []

    # Store image summaries
    image_summaries = []

    # Prompt
    prompt = """You are an assistant tasked with summarizing images for retrieval. \
    These summaries will be embedded and used to retrieve the raw image. \
    Give a concise summary of the image that is well optimized for retrieval."""


    base64_image = encode_image(path)
    img_base64_list.append(base64_image)
    image_summaries.append(image_summarize(base64_image, prompt))

    return img_base64_list, image_summaries
```

```python
import uuid
from langchain_core.documents import Document
from langchain_core.stores import InMemoryStore
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_core.runnables import RunnableLambda
```

```python
def create_manual_multi_vector_retriever(vectorstore, text_summaries, texts, table_summaries, tables, image_summaries, images):
    """
    Ingests data into a manual Multi-Vector setup and returns a Runnable retriever.
    """

    # 1. Initialize storage
    docstore = InMemoryStore()
    id_key = "doc_id"

    # 2. Define Helper (Fixed with Length Checks)
    def add_data_to_stores(summaries, raw_contents, data_type):
        if not summaries or not raw_contents:
            return

        # SAFETY CHECK: Ensure input is actually a list, not a single string
        if isinstance(summaries, str):
            print(f"Warning: {data_type} summary was passed as a string. Converting to single-item list.")
            summaries = [summaries]

        if len(summaries) != len(raw_contents):
            print(f"CRITICAL WARNING for {data_type}: Summary count ({len(summaries)}) does not match Raw Content count ({len(raw_contents)}). Truncating to shorter length.")
            # Determine the minimum length to avoid IndexError
            min_len = min(len(summaries), len(raw_contents))
            summaries = summaries[:min_len]
            raw_contents = raw_contents[:min_len]

        # Generate IDs based on the synchronized length
        doc_ids = [str(uuid.uuid4()) for _ in range(len(summaries))]

        # Create Documents for VectorStore
        summary_docs = [
            Document(page_content=s, metadata={id_key: doc_ids[i]})
            for i, s in enumerate(summaries)
        ]

        vectorstore.add_documents(summary_docs)

        # Store Raw Content
        raw_docs = [
            Document(page_content=str(content), metadata={"type": data_type})
            for content in raw_contents
        ]
        docstore.mset(list(zip(doc_ids, raw_docs)))

    # 3. Process data
    print("Ingesting texts...")
    add_data_to_stores(text_summaries, texts, "text")

    print("Ingesting tables...")
    add_data_to_stores(table_summaries, tables, "table")

    print("Ingesting images...")
    # THIS IS THE CRITICAL PART FOR YOUR ERROR
    add_data_to_stores(image_summaries, images, "image")

    # 4. Define Retrieval Logic
    def retrieve_process(query: str):
        sub_docs = vectorstore.similarity_search(query)
        ids = []
        for d in sub_docs:
            if id_key in d.metadata:
                ids.append(d.metadata[id_key])

        unique_ids = list(set(ids))
        if unique_ids:
            return docstore.mget(unique_ids)
        return []

    return RunnableLambda(retrieve_process)

# --- usage ---

# Initialize your VectorStore
vectorstore = Chroma(
    collection_name="mm_rag",
    embedding_function=OpenAIEmbeddings(
        api_key=getenv('OPENROUTER_API_KEY'),
        base_url="https://openrouter.ai/api/v1",
        model="text-embedding-3-small"
    )
)

# Create the retriever
# Ensure variable names match what you pass (e.g., 'Text' vs 'texts')
retriever = create_manual_multi_vector_retriever(
    vectorstore=vectorstore,
    text_summaries=text_summaries,
    texts=Text,
    table_summaries=table_summaries,
    tables=tab,
    # USE THE LIST OF SUMMARIES, NOT THE FINAL SUMMARY STRING
    image_summaries=all_img_summaries,
    images=image_paths
)

# Test it
# results = retriever.invoke("What are the Q3 financial results?")
```

```python
import base64
import io
import re
from PIL import Image
from langchain_core.messages import HumanMessage
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from os import getenv

# --- Helper Functions ---

def looks_like_base64(sb):
    """Check if the string looks like base64"""
    if not isinstance(sb, str): return False
    # minimal check to avoid false positives on regular text
    if len(sb) < 100: return False
    return re.match("^[A-Za-z0-9+/]+[=]{0,2}$", sb) is not None

def is_image_data(b64data):
    """Check if the base64 data is an image by looking at the header"""
    image_signatures = {
        b"\xFF\xD8\xFF": "jpg",
        b"\x89\x50\x4E\x47\x0D\x0A\x1A\x0A": "png",
        b"\x47\x49\x46\x38": "gif",
        b"\x52\x49\x46\x46": "webp",
    }
    try:
        header = base64.b64decode(b64data)[:8]
        for sig, _ in image_signatures.items():
            if header.startswith(sig):
                return True
        return False
    except Exception:
        return False

def resize_base64_image(base64_string, max_width=1000):
    """
    Resize an image encoded as a Base64 string, maintaining aspect ratio.
    Fixed: Prevents distortion (squashing/stretching).
    """
    img_data = base64.b64decode(base64_string)
    img = Image.open(io.BytesIO(img_data))

    # Only resize if the image is wider than max_width
    if img.width > max_width:
        w_percent = (max_width / float(img.size[0]))
        h_size = int((float(img.size[1]) * float(w_percent)))
        img = img.resize((max_width, h_size), Image.LANCZOS)

    buffered = io.BytesIO()
    # Save as original format or JPEG if unknown
    fmt = img.format if img.format else 'JPEG'
    img.save(buffered, format=fmt)
    return base64.b64encode(buffered.getvalue()).decode("utf-8")

def split_image_text_types(docs):
    """Split base64-encoded images and texts"""
    b64_images = []
    texts = []

    for doc in docs:
        # Handle LangChain 'Document' objects
        content = doc.page_content if hasattr(doc, "page_content") else doc

        # Check if it is an image
        if looks_like_base64(content) and is_image_data(content):
            # Resize safely maintaining aspect ratio
            resized = resize_base64_image(content)
            b64_images.append(resized)
        else:
            texts.append(str(content))

    return {"images": b64_images, "texts": texts}

def img_prompt_func(data_dict):
    """Join the context into a single prompt"""
    formatted_texts = "\n".join(data_dict["context"]["texts"])
    messages = []

    # 1. Add Images (if any)
    if data_dict["context"]["images"]:
        for image in data_dict["context"]["images"]:
            messages.append({
                "type": "image_url",
                "image_url": {"url": f"data:image/jpeg;base64,{image}"},
            })

    # 2. Add Text Prompt
    text_content = (
        "You are a helpful assistant.\n"
        "You will be given mixed information (text, tables, images).\n"
        "Use this information to answer the user question.\n"
        f"User Question: {data_dict['question']}\n\n"
        "Context (Text/Tables):\n"
        f"{formatted_texts}"
    )

    messages.append({"type": "text", "text": text_content})
    return [HumanMessage(content=messages)]

# --- Chain Definition ---

def multi_modal_rag_chain(retriever):
    """Multi-modal RAG chain"""

    # Fixed: Use gpt-4o and OpenRouter Base URL
    model = ChatOpenAI(
        temperature=0,
        model="xiaomi/mimo-v2-flash:free",
        max_tokens=1024,
        api_key=getenv('OPENROUTER_API_KEY'),
        base_url="https://openrouter.ai/api/v1"
    )

    chain = (
        {
            "context": retriever | RunnableLambda(split_image_text_types),
            "question": RunnablePassthrough(),
        }
        | RunnableLambda(img_prompt_func)
        | model
        | StrOutputParser()
    )

    return chain

# Create RAG chain using the variable name from your previous step
# (Assuming the variable name was 'retriever')
chain_multimodal_rag = multi_modal_rag_chain(retriever)
```
